#Merging the Dataframes

We merge the above dataframes into one dataframe on which EDA, Feature Enginering etc. that is, all the necessary steps will be performed, before feeding it to the Machine Learning Model

```python
#Since the results dataframe doesn't have a Driver ID column, we will create it

#Getting the driver_list from the laps dataframe
driver_list = df_laps['Driver ID'].unique()

#Extracting the last name from the driver names and creating a new feature to store
df_results['Driver ID'] = df_results['Driver Name'].str.lower().str.split().str[-1]

#Ensure that the Driver ID matches the format of Driver ID in df_laps
df_results['Driver ID'] = df_results['Driver ID'].apply(lambda x: x if x in driver_list else None)

#Check the result
df_results

```

```python
#Step 1 : Merge df_laps and df_results on 'Race Name', 'Year' and 'Driver ID'
df_combined = pd.merge(df_laps, df_results, on=['Race Name', 'Year' ,'Driver ID'],how = 'left')

#Step 2 : Merge the combined dataframe with df_pitstops on 'Race Name', 'Year' and 'Driver ID'
df_combined = pd.merge(df_combined, df_pitstops, on = ['Race Name','Year','Driver ID'],how = 'left')

#Step 3 : Merge the combined dataframe with df_weather on 'Race Name' and 'Year'
df_combined = pd.merge(df_combined, df_weather, on = ['Race Name','Year'], how='left')

#Display the first few rows of the merged dataframe to check
df_combined
```

```python
#Getting the number of pit stops for each driver
pit_stops = df_combined.groupby(['Year','Race Name','Driver ID'])['Stop'].transform('nunique')

#Broadcasting the data
df_combined['Pit Stops'] = pit_stops

#Getting the number of pit stops for each driver
df_combined['Pit Stops'].value_counts()
```
